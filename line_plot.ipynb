```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('san_antonio.csv', low_memory = False)
df.head()
```

```python
df.shape
```

```python
#dropping raws with NaN in important columns
df.dropna(subset=['raw_custodial_arrest_made','raw_row_number', 'date', 'time', 'type'], inplace=True)
df.shape
```

```python
# Splitting date column
df['date'] = pd.to_datetime(df['date'])
df['month_year'] = df['date'].dt.to_period('M').astype(str)
df.head()
```

```python
df = df.sort_values(by='month_year', ascending=True)
df
```

```python
graph_df = df.groupby('month_year').size().reset_index(name='stop_count')
graph_df.head()
```

```python
fig = px.line(graph_df, x='month_year', y='stop_count', 
              title='Monthly Police Stops in San Antonio (Jan. 2012 - Mar. 2020)',
              labels={'month_year': 'Month', 'stop_count': 'Total Stops That Month'})
fig.show()
```
